```python
# =====================================================
# TASK 1: DATA CLEANING AND PREPROCESSING
# Dataset: Stock Prices Dataset (2_Stock_Prices_Data_Set.csv)
# Tools: Python, pandas
# =====================================================
```

```python
# STEP 1: IMPORT LIBRARIES
# We start by importing pandas, which is the main library used for data manipulation and analysis in Python

import pandas as pd
```

```python
# STEP 2: LOAD THE DATASET
# We load the dataset directly from GitHub using its raw URL.

df = pd.read_csv("https://raw.githubusercontent.com/clementafangama-gif/Codveda-Data-Analytics_Internship_Datasets/refs/heads/main/2)%20Stock%20Prices%20Data%20Set.csv")

# Preview the first 5 rows of the datasets
print("First 5 rows of the dataset:")
print(df.head())
```

```python
# STEP 3: UNDERSTAND THE DATASET
''' Before cleaning, we explore the dataset to understand its structure;
how many rows, columns, and what data types are present. '''

print("\nDataset Shape (Rows, Columns):", df.shape)
print("\nColumn Names:\n", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
```

```python
# STEP 4: IDENTIFY AND HANDLE MISSING VALUES
''' Missing values can affect our analysis,
so we identify which columns have empty values, and how many. '''

print("\nMissing Values Per Column:")
print(df.isnull().sum())

print("\nTotal Missing Values:", df.isnull().sum().sum())
```

```python
# STEP 5: HANDLING MISSING VALUES
''' We fill the missing values in numerical columns with the mean (average)
of each column. This is to ensure that no data is lost, and also keep
the values within a reasonable range. '''

for col in ['open','high','low']:
    mean_value = df[col].mean()
    df[col].fillna(mean_value, inplace=True)
    print(f"Filled missing values in '{col}' with mean: {mean_value:.2f}")

# Confirm no missing values remain
print("\nMissing Values After Handling:")
print(df.isnull().sum())
print("\nTotal Missing Values Remaining:", df.isnull().sum().sum())
```

```python
# STEP 6: HANDLING DUPLICATE ROWS
# Duplicate rows can skew our analysis and models. So, we identify and remove them.

print("Number of Duplicate Rows:", df.duplicated().sum())

# Remove duplicates if any
df.drop_duplicates(inplace=True)
print("Duplicate Rows After Removal:", df.duplicated().sum().sum())
print("New Dataset Shape:", df.shape)
```

```python
# STEP 7: STANDARDIZE THE DATE COLUMN
''' The 'date' column is currently stored as a string (object) data type.
So we have to convert it to a proper datetime format so it can be used
correctly in time-based analysis such as sorting, filtering, and plotting. '''

# Convert date column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Confirm the change
print("Date Column data type:", df['date'].dtype)
print("\nSample of date column:")
print(df['date'].head())
```

```python
# STEP 8: STANDARDIZE COLUMN NAMES
''' The may be inconsistencies in column names spacing or capitalization. So we
have to clean them up and change them to all lowercase with underscores instead
of spaces, to make them easier to work with in code. '''

# Current Column Names
print("Column Names:")
print(df.columns.tolist())

# Standardize Column Names
df.columns = df.columns.str.strip().str.lower().str.replace(' ','_')

# Display Standardized Columns
print("\nColumn Names After Standardization:")
print(df.columns.tolist())
```

```python
# STEP 9: FINAL SUMMARY
''' We have to do a final check to confirm the dataset is clean and ready
for analysis. This summarizes all the cleaning steps and the final state
of the dataset. '''

print("======FINAL SUMMARY======")
print("\nFinal Dataset Summary:", df.head)
print("\nData Types:", df.dtypes)
print("Missing Values:", df.isnull().sum().sum())
print("Duplicate Rows:", df.duplicated().sum())
print("\nSample of Cleaned Dataset:")
print(df.head())
```

```python
# STEP 10: SAVE THE CLEANED DATASET
''' We have to save the cleaned dataset as a new CSV file so that it can be used in
subsequent tasks or analysis without repeating the cleaning process. Set the index to
False in order to avoid saving the row numbers as a separate column. '''

df.to_csv('Stock_Prices_Cleaned.csv', index=False)
```
