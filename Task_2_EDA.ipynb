```python
# =====================================================
# TASK 2: EXPLORATORY DATA ANALYSIS (EDA)
# Dataset: Iris Dataset (1_iris.csv)
# Tools: Python, pandas, matplotlib, seaborn
# =====================================================
```

```python
# STEP 1: IMPORT LIBRARIES
'''We import pandas for data manipulation, matplotlib and seaborn for data visualization,
and scipy for calculating statistical measures like mode'''

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# STEP 2: LOAD THE DATASET
# We load the dataset directly from GitHub using its raw URL.

df = pd.read_csv('https://raw.githubusercontent.com/clementafangama-gif/Codveda-Data-Analytics_Internship_Datasets/refs/heads/main/1)%20iris.csv')

# Preview first 5 rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())
```

```python
# STEP 3: CALCULATING SUMMARY STATISTICS
'''We begin by calculating the mean, median, mode, and standard deviation of each numerical column.
To help us understand the central tendency and spread of the data.'''

print("\nMEAN:")
print(df.mean(numeric_only=True))
print("\nMEDIAN:")
print(df.median(numeric_only=True))
print("\nMODE:")
print(df.mode(numeric_only=True).iloc[0])
print("\nSTANDARD DEVIATION:")
print(df.std(numeric_only=True))
print("n\FULL SUMMARY STATISTICS:")
print(df.describe())
```

```python
# STEP 4: VISUALIZE DATA DISTRIBUTIONS
# We plot histograms, boxplots, and scatter plots to visually understand how the data is distributed, and also identify outliers.

# HISTOGRAMS
# Histograms show the frequency distribution of each numerical column, and help us see the shape of the data.
plt.figure(figsize=(12, 12))
df.hist(bins=20, color='blue', edgecolor='black')
plt.suptitle('HISTOGRAMS', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# BOX PLOTS
# Boxplots help us identify the spread of data and outliers.
plt.figure(figsize=(12,6))
sns.boxplot(data=df, color='blue')
plt.suptitle('BOX PLOTS', fontsize=16)
plt.xlabel('Features', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.show()
```

```python
# SCATTER PLOTS
#Scatter plots visualizes the relationship between two numerical features to identify patterns across different classes.
sns.pairplot(df, hue='species', palette='Set1')
plt.suptitle('SCATTER PLOTS', fontsize=16)
plt.show()
```

```python
# FIND CORRELATION BETWEEN NUMERICAL FEATURES
# Correlation measures how strongly two numerical features are related to each other.
# A value close to 1 is a strong positive correlation, close to -1 is a strong negative correlation, and close to 0 is no correlation.

print("CORRELATION MATRIX")
print(df.corr(numeric_only=True))

# HEATMAP
# A heatmap gives a visual representation of the correlation matrix.
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='RdYlGn')
plt.title('HEATMAP', fontsize=16)
plt.show()
```
